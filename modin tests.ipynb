```python
from datetime import datetime
from dask.distributed import Client
client = Client(memory_limit='16GB', processes=False, silence_logs='error')
import os 
os.environ["MODIN_ENGINE"] = "dask"
import modin.pandas as mp
import glob
def modin_run(file_name):
    
    file_vals = mp.read_csv(file_name, sep=",", header=0)
    file_vals = file_vals.groupby("key").transform('mean') 
    print(file_vals.head(5))

    del file_vals

for filename in glob.glob('./data/*.csv'):
    print(filename)
    %timeit modin_run(filename)
```
